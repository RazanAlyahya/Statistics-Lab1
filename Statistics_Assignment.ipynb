## Q3: Prove your previous result using mean, median and mode.

```python
print("AGE")
print("The Mean is", Data['AGE'].mean())
print("The Median is", Data['AGE'].median())
ModeDIS=3*Data['AGE'].median()-2*Data['AGE'].mean()
print("The Mode is", ModeDIS)
print("Negatively skewed frequency distribution, the mean is always lesser than median and the median is always lesser than the mode.")
```

```python
print("DIS")
print("The Mean is", Data['DIS'].mean())
print("The Median is", Data['DIS'].median())
ModeDIS=3*Data['DIS'].median()-2*Data['DIS'].mean()
print("The Mode is", ModeDIS)
print("Positively skewed frequency distribution, the mean is always greater than median and the median is always greater than the mode.")
```

```python
print("RM")
print("The Mean is", Data['RM'].mean())
print("The Median is", Data['RM'].median())
ModeDIS=3*Data['RM'].median()-2*Data['RM'].mean()
print("The Mode is", ModeDIS)
print("Frequency distribution graph has a symmetrical frequency curve, then mean, median and mode will be equal")
```

## Q4: Use Skew function of Pandas to print skewness of each feature.

```python
# find skewness in each col
Data.skew(axis = 0, skipna = True)
```

- A skewness value of 0 in the output denotes a symmetrical distribution of values.
- A negative skewness value in the output indicates an asymmetry in the distribution corresponding  and the tail is larger towards the left hand side of the distribution.
 A positive skewness value in the output indicates an asymmetry in the distribution corresponding and the tail is larger towards the right hand side of the distribution.

## Q5: Use numpy to draw normal distribution.

```python
# creating a normal distribution data
normal= np.random.normal(Data['CRIM'].mean(), Data['CRIM'].std(), 1000)
count, bins, ignored = plt.hist(normal, 30, density=True)
plt.plot(bins, 1/(Data['CRIM'].std() * np.sqrt(2 * np.pi)) *np.exp( - (bins - Data['CRIM'].mean())**2 / (2 * Data['CRIM'].std()**2) ),linewidth=2, color='r')
plt.title("Normal Distribution")
plt.show()
```

## Q6: Use numpy to draw uniform distribution.

```python
s = np.random.uniform(-1,0,1000)
count, bins, ignored = plt.hist(s, 15, density=True)
plt.plot(bins, np.ones_like(bins), linewidth=2, color='r')
plt.title("Uniform Distribution")
plt.show()
```

## Q7: Use numpy to draw binomial distribution.

```python
n, p = 10, .5  
#n is specify number of trials, must be >= 0, 
#p is specify probability of success in each trial, must be in range [0, 1] 
sample = np.random.binomial(n, p, 1000)

plt.hist(sample, edgecolor='White') 
plt.title("Binomial Distribution") 
plt.show()
plt.show()


```

## Q8: Simulate Two Coin Flips, then draw the result distribution.

```python

from random import randint

results = []
heads=0
tails=0

try:
    number=int(input('how many coin tosses are you going to do:'))
    for i in range(number):
        play=randint(0,1)
        results.append(play)
        if play ==0:
            tails+=1
        else:
            heads+=1
    print('Heads:', heads)
    print('Tails:', tails)
    plt.hist(results) 
except:
    print('You have entered an invalid value.')
    
    
```
